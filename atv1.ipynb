```python

# ## 1. Importação das Bibliotecas

import numpy as np
import matplotlib.pyplot as plt

# Configuração estética dos gráficos
plt.style.use('seaborn-v0_8-whitegrid')
```

```python

# ## 2. Definição de Funções Matemáticas

# Função de 1º grau
def f1(x, a=2, b=1):
    return a*x + b

# Função de 2º grau
def f2(x, a=1, b=-4, c=3):
    return a*x**2 + b*x + c

# Função polinomial (exemplo para IA — aproximação não linear)
def f3(x):
    return 0.5*x**3 - 3*x + 1

# Função de ativação (comum em IA)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

```

```python
# ## 3. Intervalo de valores para análise

x = np.linspace(-10, 10, 400)

# ## 4. Plot 1 — Funções polinomiais e interpretação geométrica

y1 = f1(x)
y2 = f2(x)
y3 = f3(x)

plt.figure(figsize=(10,5))
#plt.plot(x, y1, label="f1(x) = 2x + 1", color="blue")
#plt.plot(x, y2, label="f2(x) = x² - 4x + 3", color="red")
plt.plot(x, y3, label="f3(x) = 0.5x³ - 3x + 1", color="green")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Comparação de Funções — Crescimento Linear, Quadrático e Cúbico")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

# ---
# ### Interpretação:
# - f1(x): representa uma relação linear, típica de regressões simples.
# - f2(x): demonstra comportamento não-linear (parábola).
# - f3(x): aproxima relações mais complexas — ideia central em redes neurais profundas,
#   onde cada camada aplica uma função não-linear semelhante.
```

```python

# ## 5. Plot 2 — Funções de Ativação (IA)
# As funções de ativação definem como os neurônios artificiais "reagem" aos estímulos.

# ReLU (Rectified Linear Unit)
def relu(x):
    return np.maximum(0, x)

# Tangente hiperbólica
def tanh(x):
    return np.tanh(x)

# Cálculo das funções de ativação
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

plt.figure(figsize=(10,5))
plt.plot(x, y_sigmoid, label="Sigmoid", color="blue")
plt.plot(x, y_relu, label="ReLU", color="orange")
plt.plot(x, y_tanh, label="tanh", color="green")
plt.title("Funções de Ativação em Inteligência Artificial")
plt.xlabel("Entrada (x)")
plt.ylabel("Saída f(x)")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.legend()
plt.show()

# ---
# ### Interpretação:
# - **Sigmoid:** suaviza valores e é usada para decisões binárias (entre 0 e 1).
# - **ReLU:** ativa apenas valores positivos — muito usada em redes neurais modernas.
# - **tanh:** centraliza valores entre -1 e 1, útil para normalização em camadas ocultas.

```
